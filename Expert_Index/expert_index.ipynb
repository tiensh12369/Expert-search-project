```python
import datetime

def recentness(pYears):
    dt = datetime.datetime.now()
    rct = 0
    lct = 0
    year_avg = sum(pYears) / len(pYears)
    if year_avg >= int(dt.year)-2:
        lct = 1
    elif int(dt.year)-15 < year_avg <= int(dt.year)-3:
        lct = max(round((1-(int(dt.year)-3-year_avg)*0.1),2), 0)

    rct += len(pYears)

    if len(pYears) == 0:
        rct = 0
        lct = 0

    return rct, lct


print(recentness([2019]))
```
